```python
from getpass import getpass

openai_key = getpass("Enter OpenAI API key:")
```

# Load data

```python
# We'll use the Wikipedia pages of Seven Wonders of the Ancient World as Documents.
# We preprocessed the data and uploaded to a Hugging Face Space: Seven Wonders.
from datasets import load_dataset
dataset = load_dataset("bilgeyucel/seven-wonders", split="train")
```

# Create document store

```python
from haystack.document_stores import FAISSDocumentStore
document_store = FAISSDocumentStore()
document_store.write_documents(dataset)
```

# Create Retriever

```python
from haystack.nodes import EmbeddingRetriever, PromptNode


retriever = EmbeddingRetriever(document_store=document_store,
                               embedding_model="sentence-transformers/multi-qa-mpnet-base-dot-v1")
```

# Define LLM

```python
from haystack.nodes import PromptTemplate, PromptNode

prompt_text = """
Answer the following question.
Question: {query}
Answer:
"""

prompt_template = PromptTemplate(prompt=prompt_text)

prompt_node = PromptNode(
    model_name_or_path = "gpt-3.5-turbo",
    default_prompt_template=prompt_template,
    api_key=openai_key,
    max_length=768,
    model_kwargs={"stream": True},
)
```

```python
prompt_node.run("Write a short poem on AI")
```

# Create a pipeline

```python
from haystack.pipelines import Pipeline

pipe = Pipeline()
pipe.add_node(component=retriever, name="retriever", inputs=["Query"])
pipe.add_node(component=prompt_node, name="prompt_node", inputs=["retriever"])
```

```python
output = pipe.run(query="What does Rhodes Statue look like?")
```
